```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
valid_losses = pd.read_csv('logs/05_optimized_scratch_val_loss.csv')
```

```python
cycle = plt.rcParams['axes.prop_cycle'].by_key()['color']
```

```python
fig, ax = plt.subplots(figsize=(9, 6))
run_no = 0

for col in valid_losses.columns: 
    if col.endswith('valid_loss'):
        run_no += 1
        plt.plot(valid_losses['epoch'], valid_losses[col], label=col.split()[0], linewidth=2, linestyle='-', color=cycle[run_no-1])
plt.xticks(np.arange(1, 26))   
plt.xlabel('epoch', fontsize='x-large')
plt.ylabel('validation loss', fontsize='x-large')
plt.ylim([-0.1, 7])
plt.legend(fontsize='large')

fig.tight_layout()

fig.savefig('../figs/glitches/05_optimized_val_loss_curves.png', dpi=300)
```
